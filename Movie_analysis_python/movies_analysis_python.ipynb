#### Import the libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configurations of the plot we'll create
```

#### Load the data from the csv file

```python
df = pd.read_csv(r'C:\Users\kagwi\OneDrive\Desktop\Portfolio_Projects\movies_correlation_project\movies.csv')

df.head()
```

#### Check for any missing data

```python
for col in df.columns:
    perct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, perct_missing))
```

#### drop the rows with null values

```python
df2 = df.dropna()
for col in df2.columns:
    perct_missing = np.mean(df2[col].isnull())
    print('{} - {}%'.format(col, perct_missing))
```

```python
df.dtypes
```

#### Change the data type of some columns

```python
df2['budget'] = df2['budget'].astype('int64')
df2['gross'] = df2['gross'].astype('int64')
```

#### Get the correct release year

```python
df2['correct_year'] = df2['released'].str.split(',').str[1]
df2['correct_year'] = df2['correct_year'].str.split('(').str[0]
```

```python
pd.set_option('display.max_rows', 10)
df2.sort_values(by=['gross'],ascending=False)
```

#### Drop any duplicates

```python
df2.drop_duplicates()
```

#### Scatter plot of budget vs gross

```python
plt.scatter(x=df2['budget'], y=df2['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

#### Plot budget vs gross using seaborn

```python
sns.regplot(x='budget', y='gross', data=df2, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

#### Check for correlation in the data 

```python
correlation_matrix = df2.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Numerical Features')
plt.ylabel('Numerical Features')
plt.show()
```

#### Numerize all the fields in the dataset so that they can be included in the correlation matrix 

Drop the year column to remain with correct_year column first

```python
del df2['year']
```

```python
df_numerized = df2
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Movie Data')
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```

#### Find the features that have the highest correlation to the gross earnings 

```python
corr_pairs = correlation_matrix.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_correlation = sorted_pairs[(sorted_pairs) > 0.5]
high_correlation
```

#### Insights 

According to the analysis, it is clear that the gross earnings are highly correlated to the budget and votes.
